Iremos fazer uma analise preditiva com base no KNN algoritmo de classificação, para isso iremos splitar os dados numa proporçaõ 8 para treino e 2 para teste

```python
x_train, x_test, y_train, y_test = train_test_split(breast_cancer_data.data, breast_cancer_data.target, train_size = 0.8, test_size = 0.2, random_state = 100)
training_data = x_train
validation_data = x_test
training_labels = y_train
validation_labels = y_test
```

Agora queremos descobrir qual o "K" que trará a maior acurácia para o teste, diminuindo a chance de erro

```python
accuracies = []
for k in range (1,100):
  classifier = KNeighborsClassifier(n_neighbors = k)
  classifier.fit(training_data, training_labels)
  accuracies.append(classifier.score(validation_data, validation_labels))
k_list = range(1,100)
```

```python
plt.plot(k_list, accuracies)
plt.xlabel("Possíveis k's")
plt.ylabel("Acurácio do teste para cada k")
plt.title("Gráfico Acurácia x K")
plt.show()
```

É possível perceber que o k aproximadamente 55 ou k igual a 20 resultam nos teste com menor taxa de erro.
